##### Calculating Race/Ethnicity & Employment Percentages

```python
PercWhite = (cookbgdemo['White']/cookbgdemo['TotalAll']) * 100
PercBlack = (cookbgdemo['BlackorAA']/cookbgdemo['TotalAll']) * 100
PercNative = (cookbgdemo['Native']/cookbgdemo['TotalAll']) * 100
PercAsian = (cookbgdemo['Asian']/cookbgdemo['TotalAll']) * 100
PercPIslander = (cookbgdemo['PacificIslander']/cookbgdemo['TotalAll']) * 100
PercOther = (cookbgdemo['Other']/cookbgdemo['TotalAll']) * 100
PercTwoPlus = (cookbgdemo['TwoPlus']/cookbgdemo['TotalAll']) * 100
PercHispLatino = (cookbgdemo['HispanicLatino']/cookbgdemo['TotalAll']) * 100
PercEmployed = (cookbgdemo['TotalCivEmployed']/cookbgdemo['TotalCivLaborForce']) * 100
PercUnemployed = (cookbgdemo['TotalCivUnemployed']/cookbgdemo['TotalCivLaborForce']) * 100
```

##### Adding Race/Ethnicity & Employment Percentages to dataframe

```python
cookbgdemo['PercWhite'] = PercWhite
cookbgdemo['PercBlack'] = PercBlack
cookbgdemo['PercNative'] = PercNative
cookbgdemo['PercAsian'] = PercAsian 
cookbgdemo['PercPacificIslander'] = PercPIslander 
cookbgdemo['PercOther'] = PercOther 
cookbgdemo['PercTwoPlus'] = PercTwoPlus 
cookbgdemo['PercHispanicLatino'] = PercHispLatino 
cookbgdemo['PercCivEmployed'] = PercEmployed
cookbgdemo['PercCivUnemployed'] = PercUnemployed
```

##### Calc Educational Attainment Percentages 

```python
PercNoHS = ((cookbgdemo['NoSchooling'] + cookbgdemo['Nursery'] + cookbgdemo['Kindergarten'] + cookbgdemo['1stGrade'] 
+ cookbgdemo['2ndGrade'] + cookbgdemo['3rdGrade'] + cookbgdemo['4thGrade'] + cookbgdemo['5thGrade'] 
+ cookbgdemo['6thGrade'] + cookbgdemo['7thGrade'] + cookbgdemo['8thGrade'] + cookbgdemo['9thGrade'] 
+ cookbgdemo['10thGrade'] + cookbgdemo['11thGrade'] + cookbgdemo['12thGradeNoGrad']) / cookbgdemo['TotalEdu']) * 100

PercHS = ((cookbgdemo['HSGrad'] + cookbgdemo['GEDHSGrad'])/cookbgdemo['TotalEdu']) * 100

PercCollege = ((cookbgdemo['1YrCollege'] + cookbgdemo['1PlusYrsCollegeNoDeg'] + cookbgdemo['AssociatesDeg']
                + cookbgdemo['BachelorsDeg'])/cookbgdemo['TotalEdu']) * 100

PercMasters = (cookbgdemo['MastersDeg']/cookbgdemo['TotalEdu']) * 100

PercPro = (cookbgdemo['ProSchoolDeg']/cookbgdemo['TotalEdu']) * 100

PercDoctorate = (cookbgdemo['DoctorateDeg']/cookbgdemo['TotalEdu']) * 100
```

##### Adding Educational Attainment Percentages to dataframe

```python
cookbgdemo['PercNoHS'] = PercNoHS
cookbgdemo['PercHS'] = PercHS
cookbgdemo['PercCollege'] = PercCollege
cookbgdemo['PercMasters'] = PercMasters
cookbgdemo['PercProSchool'] = PercPro
cookbgdemo['PercDoctorate'] = PercDoctorate
```
